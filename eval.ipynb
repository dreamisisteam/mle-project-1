```python
import os
import shutil

from ultralytics import YOLO
from PIL import Image
```

```python
TEST_DATA_DIR = "./test_data"
PREDICTED_DATA_DIR = "./pred_data"
TEST_IMAGES = [TEST_DATA_DIR + "/" + image_name for image_name in os.listdir(TEST_DATA_DIR)]
```

```python
TEST_IMAGES
```

```python
model = YOLO("./models/best.pt")
```

```python
result_images = model(TEST_IMAGES, conf=0.5)
```

```python
for image in result_images:
    
    if not os.path.exists(PREDICTED_DATA_DIR):
        os.mkdir(PREDICTED_DATA_DIR)
    image_name = image.path.split("/")[-1].split(".")[0]

    image.show()
    image.save(filename=PREDICTED_DATA_DIR + "/" + image_name + ".jpg")
```
